```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].str.lower()
carreteras_federales
```

```python
import unicodedata
import re
```

```python
def limpiar_texto(texto):
    # Quitar acentos
    texto = ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn' or c == 'ñ'
    )
    texto = unicodedata.normalize('NFC', texto)  # Normalizar de nuevo a forma compuesta

    # Eliminar caracteres especiales
    texto = re.sub(r'[^a-zA-ZñÑ\s]', '', texto)



    return texto
```

```python
def quitar_acentos(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')
    # Para mantener Ñ
    #def quitar_acentos(texto):
    #texto = ''.join(
     #   c for c in unicodedata.normalize('NFD', texto)
      #  if unicodedata.category(c) != 'Mn' or c == 'ñ'
    #)
    #return unicodedata.normalize('NFC', texto)
```

```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].apply(limpiar_texto)
```

```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].str.replace(' ', '')
```

```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].str.replace('?', '')
```

```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].str.replace('-', '')
```

```python
carreteras_federales['NOMBRE'] = carreteras_federales['NOMBRE'].str.replace('.', '')
```

```python
# para_disolver= ['NOMBRE', 'CODIGO']
carreteras_agrupadas= carreteras_federales.dissolve(by=['NOMBRE', 'CODIGO'], aggfunc={
    'ID_RED':'count',
    'LONGITUD':'sum',
})
```

```python
carreteras_agrupadas
```

```python
info_carreteras = carreteras_agrupadas.drop(['geometry'], axis=1)
```

```python
# prompt: cambaiar el text de la columna ID_RED	a NUM_Tramos

info_carreteras = info_carreteras.rename(columns={'ID_RED': 'NUM_Tramos'})
info_carreteras
```

```python
# prompt: agregar una columna que contenga el nombre

info_carreteras['Nombre'] = info_carreteras.index.get_level_values('NOMBRE')
info_carreteras
```
