```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aguascalientes = gpd.read_file('/content/drive/MyDrive/Carreteras/sin_bam/Aguascalientes_total_carr2023_sincob1t24.shp')
```

```python
agsc_1S=aguascalientes.loc[aguascalientes['CIRCULA']="Un sentido"]
agsc_2S=aguascalientes.loc[aguascalientes['CIRCULA']="Dos sentidos"]
```

```python
agsc_1S['longitud_metros']=agsc_1S['geometry'].length
agsc_2S['longitud_metros']=agsc_2S['geometry'].length
long_1s=agsc_1S['longitud_metros'].sum()
long_1s=long_1s/2
long_2s=agsc_2S['longitud_metros'].sum()
long_total=long_1s+long_2s
print(long_total)

```

```python
def calcular_longitud_total(file_path):
    data = gpd.read_file(file_path)
    un_sentido = data.loc[data['CIRCULA'] == "Un sentido"].copy()
    dos_sentidos = data.loc[data['CIRCULA'] == "Dos sentidos"].copy()

    # Calcular la longitud en metros
    un_sentido['longitud_metros'] = un_sentido['geometry'].length
    dos_sentidos['longitud_metros'] = dos_sentidos['geometry'].length

    # Sumamos las longitudes
    long_1s = un_sentido['longitud_metros'].sum() / 2  # Un sentido, se divide entre 2
    long_2s = dos_sentidos['longitud_metros'].sum()    # Dos sentidos
    long_total = long_1s + long_2s

    return long_total
```

```python
import os

# Lista de nombres de los estados en orden alfabético
estados = [
    "Aguascalientes", "Baja California", "Baja California Sur", "Campeche",
    "Chiapas", "Chihuahua", "Coahuila de Zaragoza", "Colima", "Ciudad de México",
    "Durango", "Guanajuato", "Guerrero", "Hidalgo", "Jalisco", "México",
    "Michoacán de Ocampo", "Morelos", "Nayarit", "Nuevo León", "Oaxaca", "Puebla", "Querétaro",
    "Quintana Roo", "San Luis Potosí", "Sinaloa", "Sonora", "Tabasco", "Tamaulipas",
    "Tlaxcala", "Veracruz de Ignacio de la Llave", "Yucatán", "Zacatecas"
]

# Directorio donde están almacenados los archivos
base_path = '/content/drive/MyDrive/Carreteras/sin_bam/'

```

```python
# Almacenar los resultados
longitudes_totales = {}

for estado in estados:
    file_path = os.path.join(base_path, f'{estado}_total_carr2023_sincob1t24.shp')
    longitud = calcular_longitud_total(file_path)
    longitudes_totales[estado] = longitud
    print(f'{estado}: {longitud:}')
```

```python
longitudes_totales
```

```python
df_longitudes = pd.DataFrame(list(longitudes_totales.items()), columns=['Estado', 'Longitud (Metros)'])

# Guardar el DataFrame en un archivo Excel
output_file = '/content/drive/MyDrive/Carreteras/longitudes_totales.xlsx'
df_longitudes.to_excel(output_file, index=False)

print(f'Archivo guardado en: {output_file}')
```
