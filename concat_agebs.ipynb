```python
# prompt: guardar como csv y escoger la ruta de destino

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
!pip install  geopandas pandas matplotlib

agebs = gpd.read_file('/content/drive/MyDrive/MG_2020_Integrado/conjunto_de_datos/00a.shp')
agebs_bc = agebs[agebs['CVE_ENT']=='02']
agebs_bc
#juntar el texto de los datos de las columnas CVE_ENT , CVE_MUN y CVE_LOC y añadirlos como otra columna

agebs_bc['Localidad'] = agebs_bc['CVE_ENT'] + agebs_bc['CVE_MUN'] + agebs_bc['CVE_LOC']
agebs_bc

# agebs_bc['LOCALIDAD'] = agebs_bc[['CVE_ENT','CVE_MUN','CVE_LOC']].str.cat()#agg(''.join, axis=1)
# agebs_bc
agebs_bc_X_Localidad = agebs_bc.groupby('Localidad').agg({
    'CVE_AGEB':'nunique',
    })
agebs_bc_X_Localidad

from google.colab import files
agebs_bc_X_Localidad.to_csv('agebs_bc_X_Localidad.csv')
files.download('agebs_bc_X_Localidad.csv')
```
