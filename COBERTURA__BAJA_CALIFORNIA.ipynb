```python
agebs = gpd.read_file('/content/drive/MyDrive/MG_2020_Integrado/conjunto_de_datos/00a.shp')
agebs_bc = agebs[agebs['CVE_ENT']=='02']
agebs_bc.plot()
```

```python
agebs_bc['cobertura'] = agebs_bc['geometry'].intersects(cobertura_bc['geometry'])
agebs_bc.head()
```

```python
agebs_bc[agebs_bc['cobertura']==True].count()
```

```python
agebs_bc.crs
```

```python
cobertura_bc.crs
```

```python
agebs_bc = agebs_bc.set_crs(epsg=6372, allow_override=True)
```

```python
agebs_bc.crs
```

```python
agebs_bc.plot()
```

```python
res_intersect = agebs_bc.overlay(cobertura_bc, how='intersection', keep_geom_type=False)
res_intersect
```

```python
res_intersect['cobertura']= True
agebs_bc_con_cobertura = res_intersect
agebs_bc_con_cobertura
```

```python
axis = agebs_bc_con_cobertura.plot(color = 'green')
agebs_bc.plot(ax=axis, facecolor='none', edgecolor='k')
```

```python
agebs_bc_sin_cobertura = agebs_bc.overlay(cobertura_bc, how='difference', keep_geom_type=False)
```

```python
agebs_bc_sin_cobertura.plot()
```

```python
agebs_bc_sin_cobertura.head()
```

```python
agebs_bc_sin_cobertura['cobertura']= False
```

```python
axis = agebs_bc_sin_cobertura.plot(color = 'red')
agebs_bc.plot(ax=axis, facecolor='none', edgecolor='k')
```

```python
unir los dataframes 'agebs_bc_con_cobertura' y 'agebs_bc_sin_cobertura' en uno solo

import pandas as pd

# Assuming 'agebs_bc_con_cobertura' and 'agebs_bc_sin_cobertura' are already defined

# Concatenate the two GeoDataFrames
agebs_bc_all = pd.concat([agebs_bc_con_cobertura, agebs_bc_sin_cobertura], ignore_index=True)

# Print the resulting GeoDataFrame
print(agebs_bc_all)
```
